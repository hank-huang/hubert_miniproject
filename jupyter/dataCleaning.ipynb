```python
import pandas as pd
import matplotlib.pyplot as plt
%pylab inline
```

```python
df = pd.read_csv('../datFiles/raw/justALittleMoreEditing.csv', index_col = 0)
df.drop(df.tail(1).index,inplace=True)
df['Age@Scan'] = df['Age@Scan'].astype('float64')

print(df.ConsensusDiagnosis.value_counts())

print("\nDataType")
print(df['Age@Scan'].dtypes)
```

```python
df.head()
```

```python
ncf = df['Diagnosis@LastScan'] == 'NCF'
nco = df['Diagnosis@LastScan'] == 'NCO'

df.loc[nco | ncf , 'Diagnosis@LastScan'] = 0
df.loc[df['Diagnosis@LastScan'] != 0 , 'Diagnosis@LastScan'] = 1
```

```python
target = df.ConsensusDiagnosis
features = df.drop(['ConsensusDiagnosis', 'ScanDate', 'Intracranial_Volume', 'LEC_Thickness', 'LEC_Volume' 
                   ,'REC_Volume', 'REC_Thickness', 'StudyID'], axis = 1)
```

```python
features.to_csv('../datFiles/features.csv')
target.to_csv('../datFiles/target.csv', header = False)
```
